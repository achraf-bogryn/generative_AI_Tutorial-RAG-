## Faiss

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
docs = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=30)
docs = text_splitter.split_documents(docs)
```

```python
docs
```

```python
len(docs)
```

```python
embeddings = OllamaEmbeddings()
db = FAISS.from_documents(docs, embeddings)
```

```python
# quering
query = "How many times does the word 'freedom' appear in the speech?"
docs=db.similarity_search(query)
docs
```

```python
docs[0].page_content  
```

### As a Retriever


```python
retriever = db.as_retriever()
retriever.invoke(query)
```

```python
docs[0].page_content
```

### Similarity Search With Score

```python
docs_and_scores = db.similarity_search_with_score(query)
docs_and_scores
```

`Instedad send a sentences we'll send a vector ofnumber by applying numbers`

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
docs_score  =  db.similarity_search_with_score(embedding_vector)
docs_score
```

## Saving and Loading

```python
#  save
db.save_local("faiss_index")
```

```python
new_df = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
docs = new_df.similarity_search(query)
```
